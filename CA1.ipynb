```python
import numpy as np

x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2, 5, 6, 6, 9, 8])

# Fit polynomials
coeffs_deg1 = np.polyfit(x, y, 1)
coeffs_deg5 = np.polyfit(x, y, 5)

# Generate predictions
y_pred_deg1 = np.polyval(coeffs_deg1, x)
y_pred_deg5 = np.polyval(coeffs_deg5, x)

# Compute SSR
ssr_deg1 = np.sum((y - y_pred_deg1)**2)
ssr_deg5 = np.sum((y - y_pred_deg5)**2)

print(f"SSR (Degree 1): {ssr_deg1}")
print(f"SSR (Degree 5): {ssr_deg5}")
```

#### The SSR for the degree 5 model is effectively zero

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

x_train = np.array([1, 2, 3, 4, 5, 6])
y_train = np.array([2, 5, 6, 6, 9, 8])
x_test = np.array([7, 8, 9])
y_test = np.array([11, 12, 14])


p1 = np.polyfit(x_train, y_train, 1)
p5 = np.polyfit(x_train, y_train, 5)

x_range = np.linspace(1, 9.5, 200)
y_range1 = np.polyval(p1, x_range)
y_range5 = np.polyval(p5, x_range)

plt.figure(figsize=(10, 6))

plt.scatter(x_train, y_train, color='blue', label='Training Data')
plt.scatter(x_test, y_test, color='red', marker='X', s=100, label='Test Data')

plt.plot(x_range, y_range1, color='green', linestyle='--', label='Degree 1 (Linear)')
plt.plot(x_range, y_range5, color='orange', label='Degree 5 (Polynomial)')

plt.title('Polynomial Curve Fitting: Degree 1 vs Degree 5')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.ylim(-20, 20)  
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.show()
```
