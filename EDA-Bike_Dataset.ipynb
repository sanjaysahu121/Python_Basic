#### Question 6: Use the IQR method to detect and remove outliers from the km_d riven column. Show before-and-after summary statistics.

```python
# before removing outliers summary statistics


df.describe()
```

```python
# before removing a outliers from the 'Km_driven'
sns.boxplot(data = df,palette = 'Set3')
```

```python
def dectect_ouliers(columns):
    outliers_indices = []
    for col in columns:
        Q1 = df_copy[columns].quantile(0.25)
        Q3 = df_copy[columns].quantile(0.75)
        IQR = Q3-Q1
        # mask is a boolean condition 
        mask = (df_copy[columns]>=(Q1 - 1.5*IQR)) & (df_copy[columns]<=(Q3 + 1.5*IQR))
        mask = mask.to_numpy()    # bool to numpy array 


        # to find the indices of values that do not staisfy the mas condition -> that is a outliers
        false_indices = np.argwhere(~mask)
        outliers_indices.append(false_indices)
    return np.unique(np.concatenate(outliers_indices).ravel())
```

```python
df.info()
```

```python
num_cols = ['selling_price','year','km_driven']
num_cols
```

```python
cat_cols = ['name','seller_type','owner','ex_showroom_price']
cat_cols
```

```python
dectect_ouliers('km_driven')
df_copy.drop(dectect_ouliers('km_driven'),axis =0,inplace = True)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
oh_transformer = OneHotEncoder(drop = 'first')
preprocessor  = ColumnTransformer([
    ('OneHotEncoder',oh_transformer,['seller_type'])
    
],remainder='passthrough')
preprocessor
```

```python
x_transformed = preprocessor.fit_transform(df)
x_transformed
```

```python
seller_type = [col.split('__')[-1] for col in preprocessor.get_feature_names_out()]
df_final = pd.DataFrame(x_transformed, columns=seller_type)
df_final = df_final.apply(pd.to_numeric, errors='ignore')
df_final = df_final.convert_dtypes()
```
