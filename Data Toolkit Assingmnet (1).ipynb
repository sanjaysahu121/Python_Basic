## Theoretical Questions

#### Q1.What is NumPy, and why is it widely used in Python?

##### NumPy (Numerical Python) is an open-source library used for working with numerical data in Python. It is widely used because it provides:

- Support for large, multi-dimensional arrays and matrices.

- A collection of mathematical functions to operate on these arrays efficiently.

- Faster performance than standard Python lists due to its implementation in C and contiguous memory storage.

##### Q2.How does broadcasting work in NumPy?
  - Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes. It "stretches" the smaller array across the larger one so that they have compatible shapes for element-wise operations, without making unnecessary copies of data in memory

#### Q3.What is a Pandas DataFrame?
   - A Pandas DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). Think of it like a SQL table or an Excel spreadsheet.

#### Q4.Explain the use of the groupby() method in Pandas.
- The groupby() method is used to split data into groups based on some criteria, apply a function (like mean, sum, or count) to each group independently, and then combine the results into a data structure.

##### Q5.Why is Seaborn preferred for statistical visualizations?
- Seaborn is built on top of Matplotlib and is preferred because:

   - It provides a high-level interface for drawing attractive and informative statistical graphics.

  - It integrates closely with Pandas DataFrames.

  - It automatically handles complex visualizations like linear regression fits and heatmaps with simple commands.

#### Q6.What are the differences between NumPy arrays and Python lists?

### NumPy Array
- Homogeneous (all elements must be the same type).
- Contiguous memory allocation (more efficient).
- Much faster for mathematical operations.

### Python List
- Heterogeneous (can store different types).
- Non-contiguous (less efficient).
- Slower for large-scale math

#### Q7.What is a heatmap, and when should it be used?
- A heatmap is a graphical representation of data where individual values contained in a matrix are represented as colors. It should be used to visualize correlation matrices or to show patterns and clusters in large datasets.

#### Q8.What does the term "vectorized operation" mean in NumPy?
- Vectorization refers to the practice of performing operations on entire arrays at once rather than using explicit loops (like for loops). This is faster because the operations are pushed down into highly optimized C and Fortran code.

#### Q9.How does Matplotlib differ from Plotly?
- Matplotlib: Best for static, publication-quality 2D plots. It is very flexible but requires more code for complex designs.

- Plotly: Best for interactive web-based visualizations. It allows users to zoom, pan, and hover over data points to see details

##### Q10. What is the significance of hierarchical indexing in Pandas?
- Hierarchical indexing (MultiIndex) allows you to have multiple index levels on an axis. This is significant because it lets you represent and work with higher-dimensional data in a 2D DataFrame format

#### Q11.What is the role of Seaborn's pairplot() function?
- The pairplot() function creates a grid of Scatter plots for every numerical variable in a dataset against every other variable. It is a quick way to identify relationships and distributions across the entire dataset.

#### Q12.What is the purpose of the describe() function in Pandas?
- The describe() function provides a statistical summary of a DataFrame, including count, mean, standard deviation, minimum, maximum, and quartiles (25%, 50%, 75%)

#### Q13.Why is handling missing data important in Pandas?
- Handling missing data (NaN values) is crucial because most machine learning models cannot process them. If left unaddressed, they can lead to inaccurate statistical results or cause errors in your code.

#### Q14.What are the benefits of using Plotly for data visualization?
- Interactivity: Zooming, hovering, and filtering.

- Ease of Use: High-level API for complex 3D and geographic charts.

- Web-Ready: Easily embeddable in websites and dashboards

#### Q15.How does NumPy handle multidimensional arrays?

- NumPy uses an ndarray object. It tracks the shape, size, and dtype of the data. Multidimensional arrays are stored as a flat block of memory, but NumPy uses "strides" to understand how to move between rows, columns, and higher dimensions.

#### Q16.What is the role of Bokeh in data visualization?
- Bokeh is used for creating interactive visualizations for modern web browsers. It is similar to Plotly but offers more lower-level control for building complex, custom dashboards.

#### Q17.Explain the difference between apply() and map() in Pandas.
- map(): Used for substituting each value in a Series with another value based on a dictionary or function.
- apply(): Used to apply a function along an entire axis (row or column) of a DataFrame.


#### Q18.What are some advanced features of NumPy?
- Linear Algebra: Eigenvalues, matrix decomposition.

- Fourier Transforms: Processing signals and images.

- Random Number Generation: Sophisticated distributions for simulations

#### Q19.How does Pandas simplify time series analysis?
- Pandas provides built-in support for date ranges, frequency conversion, moving window statistics (like rolling means), and easy date-based slicing

#### Q20.What is the role of a pivot table in Pandas?
- A pivot table reshapes a DataFrame into a summary table by grouping data and applying an aggregation function, making it easy to compare multiple variables

#### Q21.Why is NumPy's array slicing faster than Python's list slicing?
- NumPy slicing creates a "view" of the original array rather than a new copy. This means it doesn't need to re-allocate memory, making it nearly instantaneous regardless of array size

#### Q22.What are some common use cases for Seaborn?
- Visualizing distributions (Histograms, KDE plots).
- Examining correlations (Heatmaps).
- Comparing categories (Box plots, Violin plots).

#### Q8.Create a line plot with multiple lines using Matplotlib.


```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]  # Line 1: Squares
y2 = [1, 8, 27, 64, 125] # Line 2: Cubes
plt.plot(x, y1, label='Squares', color='blue', marker='o')
plt.plot(x, y2, label='Cubes', color='red', marker='s')
plt.title("Multiple Line Plot")
plt.xlabel("Input Value")
plt.ylabel("Output Value")
plt.legend()
plt.show()
```

#### Q13.Create a 3D scatter plot using Plotly.

```python
import plotly.express as px
import pandas as pd
import numpy as np

data = {
    'Feature_X': np.random.rand(50),
    'Feature_Y': np.random.rand(50),
    'Feature_Z': np.random.rand(50),
    'Category': np.random.choice(['A', 'B'], 50)
}
df_3d = pd.DataFrame(data)

fig = px.scatter_3d(df_3d, x='Feature_X', y='Feature_Y', z='Feature_Z',
                    color='Category', title="3D Scatter Plot Visualization")

fig.show()
```
